```python
import geopandas as gpd
import random
import pandas as pd
import matplotlib.pyplot as plt

mapa = gpd.read_file("ProvinciasArgentina.geojson")
solucion = pd.read_csv("solucion.csv")

id_a_color = pd.Series(solucion.color.values, index=solucion.id).to_dict()

nombre_a_id = {
    "Tucumán": 23,
    "Santiago del Estero": 21,
    "Salta": 16,
    "Jujuy": 9,
    "Formosa": 8,
    "Chaco": 3,
    "Misiones": 13,
    "Corrientes": 6,
    "Catamarca": 2,
    "Córdoba": 5,
    "Santa Fe": 20,
    "Entre Ríos": 7,
    "Buenos Aires": 1,
    "La Pampa": 10,
    "La Rioja": 11,
    "San Juan": 17,
    "San Luis": 18,
    "Mendoza": 12,
    "Neuquén": 14,
    "Río Negro": 15,
    "Chubut": 4,
    "Santa Cruz": 19,
    "Tierra del Fuego": 22,
    "Capital Federal": 1,
}

n = 30  # cantidad de claves
color_dicc = {i: (random.randint(0,255)/255.0, random.randint(0,255)/255.0, random.randint(0,255)/255.0) for i in range(n)}

# Colorear provincias con colores aleatorios
mapa["colores"] = mapa["nombre"].map(nombre_a_id).map(id_a_color).map(color_dicc)
mapa.plot(column=None, color=mapa["colores"], edgecolor='black')

plt.show()
```
